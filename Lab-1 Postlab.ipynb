```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Titanic.csv")

print(df.head())
```

```python
print(df.info())
```

```python
print(df.shape)
```

```python
print(df.columns)
```

```python
print(df.isnull().sum())
```

```python
print(df.describe())
```

```python
df["Survived"].value_counts().plot(kind="bar")

plt.title("Survival Count (24EU01105)")
plt.xlabel("Survived")
plt.ylabel("Number of Passengers")

plt.show()
```

```python
df["Pclass"].value_counts().sort_index().plot(kind="bar")

plt.title("Passenger Class Distribution (24EU01105)")
plt.xlabel("Passenger Class")
plt.ylabel("Count")

plt.show()
```

```python
plt.hist(df["Age"].dropna(), bins=20)

plt.title("Age Distribution (24EU01105)")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.hist(df["Fare"], bins=20)

plt.title("Fare Distribution (24EU01105)")
plt.xlabel("Fare")
plt.ylabel("Frequency")

plt.show()
```

```python
df["Sex"].value_counts().plot(kind="pie", autopct="%1.1f%%")

plt.title("Gender Distribution (24EU01105)")
plt.ylabel("")

plt.show()
```

```python
plt.boxplot(df["Fare"])

plt.title("Box Plot of Fare (24EU01105)")

plt.show()
```

```python
corr = df.select_dtypes(include="number").corr()

print(corr)
```

```python
plt.imshow(corr, cmap="coolwarm")
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

plt.title("Correlation Heatmap (24EU01105)")

plt.show()
```
