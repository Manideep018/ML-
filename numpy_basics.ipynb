```python
import numpy as np

data = np.array([12, 15, 18, 22, 30])

print("Average value (Mean):", np.mean(data))
print("Highest value (Max):", np.max(data))
print("Standard Deviation:", np.std(data))
```

```python
# Zeros
print(np.zeros((2, 3)))

# Ones
print(np.ones((2, 3)))

# Identity matrix
print(np.eye(3))

# Range of numbers
print(np.arange(0, 10, 2))
```

```python
arr = np.array([10, 20, 30, 40, 50])

print(arr[0])      # 10
print(arr[-1])     # 50
print(arr[1:4])    # [20 30 40]
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(a + b)   # [5 7 9]
print(a - b)   # [-3 -3 -3]
print(a * b)   # [4 10 18]
print(a / b)
```

```python
# Random floats between 0 and 1
print(np.random.rand(3))

# Random integers
print(np.random.randint(1, 10, size=5))
```

```python
arr = np.array([1, 2, 3, 4])

print(arr + 10)
print(arr * 2)
```

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

result = np.concatenate((a, b))
print(result)
```

```python
arr = np.array([1, 2, 3, 4, 5, 6])

parts = np.split(arr, 3)

print(parts)
```

```python
import pandas as pd

# Example 1: Create a DataFrame from a dictionary
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35],
    "City": ["New York", "London", "Paris"]
}
df = pd.DataFrame(data)
print(df)
```

```python
# Example 2: Read a CSV file
df = pd.read_csv("sample.csv")
print(df.head())  # show first 5 rows
```

```python
# Example 3: Filtering rows
young = df[df["Age"] < 30]
print(young)
```

```python
# Example 4: Grouping and aggregation
grouped = df.groupby("City")["Age"].mean()
print(grouped)
```

```python
# Example 5: Adding a new column
df["Age in 5 years"] = df["Age"] + 5
print(df)
```
